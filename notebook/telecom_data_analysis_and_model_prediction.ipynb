```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Fetching data
data_path = os.path.join(os.pardir,"data","telecom.csv")
df = pd.read_csv(data_path)
```

```python
df.info()
```

```python
df.head()
```

```python
# Converting String data types to integer
# decision trees implemented in scikit-learn uses only numerical features 
# these features are interpreted always as continuous numeric variables.

df['International plan'] = df['International plan'].map({'Yes' : 1 , 'No' : 0 })
df['Voice mail plan'] = df['Voice mail plan'].map({'Yes': 1 , 'No' : 0 })
df['Churn'] = df['Churn'].astype('int64')
```

```python
df.head()
```

```python
states = df['State']
y = df['Churn']
df.drop(['State','Churn'],axis=1,inplace=True)
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split,StratifiedKFold
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier

X_train,X_holdout,y_train,y_holdout = train_test_split(df.values , y , test_size = 0.3, random_state = 17)
tree = DecisionTreeClassifier(max_depth = 5 , random_state = 17)
knn = KNeighborsClassifier(n_neighbors = 10)

tree.fit(X_train,y_train)
knn.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score

knn_pred = knn.predict(X_holdout)
print("Accuracy score in KNN = {}".format(accuracy_score(y_holdout,knn_pred)))
```

```python
tree_pred = tree.predict(X_holdout)
print("Accuracy score in Decision Tree = {}".format(accuracy_score(y_holdout,tree_pred)))
```

```python
# Parameter tuning for Decision Tree

from sklearn.model_selection import GridSearchCV,cross_val_score

tree_params = { 'max_depth' : range(1,11) ,
                'min_samples_leaf' : range(1,11),
                 'max_features' : range(4,19)}
tree_grid = GridSearchCV(tree,tree_params, cv = 5, n_jobs = 1,verbose = True)
tree_grid.fit(X_train,y_train)
```

```python
tree_grid.best_params_
```

```python
tree_grid.best_estimator_
```

```python
tree_grid.best_score_
```

```python
tree_grid.best_index_
```

```python
best_params_tree_predict = tree_grid.predict(X_holdout)
print("Accuracy score for Decision Tree with best params = {} ".format(accuracy_score(y_holdout,best_params_tree_predict)))
```

```python
# Parameter tuning for KNN

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

knn_pipe = Pipeline([('scaler',StandardScaler()),('knn', KNeighborsClassifier(n_jobs= 1))])
knn_params = {'knn__n_neighbors' : range(1,10)}

knn_grid = GridSearchCV(knn_pipe, knn_params, cv= 5, n_jobs= 1 , verbose=True)
knn_grid.fit(X_train,y_train)
```

```python
knn_grid.best_params_ , knn_grid.best_score_
```

```python
best_params_knn_predict = knn_grid.predict(X_holdout)
print("Accuracy score for KNN with best params = {} ".format(accuracy_score(y_holdout,best_params_knn_predict)))
```

```python
# Random Forest 
from sklearn.ensemble import RandomForestClassifier

forest = RandomForestClassifier(n_estimators = 100 ,n_jobs = -1 , random_state = 17)
print(np.mean(cross_val_score(forest , X_train , y_train , cv = 5)))
```

```python
forest_params = { 'max_depth' : range(1,11),
                  'max_features' : range(4,19),
                  'min_samples_leaf' : range(1,11)}

forest_grid = GridSearchCV(forest,forest_params, cv=5 , n_jobs = -1 , verbose = True)
forest_grid.fit(X_train,y_train)
```

```python
forest_grid.best_params_
```

```python
forest_grid.best_score_
```

```python
from xgboost import XGBClassifier

xg = XGBClassifier(n_estimators = 1000 , learning_rate = 0.5)
xg.fit(X_train,y_train)
```

```python
xg_predict = xg.predict(X_holdout)
print("Accuracy score for xgboost {}".format(accuracy_score(y_holdout,xg_predict)))
```

```python
xg_params = { 'min_child_weight': [1, 5, 10],
                'gamma': [0.5, 1, 1.5, 2, 5],
                'subsample': [0.6, 0.8, 1.0],
                'colsample_bytree': [0.6, 0.8, 1.0],
                'max_depth': [3, 4, 5]
        }
xg_grid = GridSearchCV(estimator=xg,param_grid=xg_params, cv = 5 , verbose=True)
xg_grid.fit(X_train,y_train)
```

```python
xg_grid.best_params_
```

```python
from sklearn.model_selection import RandomizedSearchCV
folds = 3
param_comb = 5

skf = StratifiedKFold(n_splits=folds, shuffle = True, random_state = 1001)

random_search = RandomizedSearchCV(xg, param_distributions=xg_params, n_iter=param_comb, scoring='roc_auc', n_jobs=1, cv=skf.split(X_train,y_train), verbose=3, random_state=1001 )
random_search.fit(X_train, y_train)
```

```python
random_search.best_params_
```
